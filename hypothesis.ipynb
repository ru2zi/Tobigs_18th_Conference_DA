### Mann-Whitney U 검정(비모수)

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import plotly.express as px

# Mann-Whitney U 검정 수행 함수
def perform_mann_whitney_tests(df, threshold_col, threshold_value):
    df[f'{threshold_col}_high'] = df[threshold_col] >= threshold_value
    group1 = df[df[f'{threshold_col}_high']]['EPDO_cnt']
    group2 = df[~df[f'{threshold_col}_high']]['EPDO_cnt']
    
    # 그룹이 비어있는지 확인
    if len(group1) == 0 or len(group2) == 0:
        return np.nan, np.nan
    
    u_stat, p_value = stats.mannwhitneyu(group1, group2)
    return u_stat, p_value

# 임계값 변화시키며 Mann-Whitney U 검정 수행 함수
def find_mann_whitney_thresholds(df, col, min_val, max_val):
    results = []
    for threshold in range(min_val, max_val + 1):
        u_stat, p_value = perform_mann_whitney_tests(df, col, threshold)
        if not np.isnan(u_stat) and not np.isnan(p_value):
            results.append((col, threshold, u_stat, p_value))
    return results

def plot_threshold_results(results, variable_name):
    thresholds = [result[1] for result in results]
    p_values = [result[3] for result in results]
    
    fig = px.scatter(x=thresholds, y=p_values, title=f'{variable_name} 임계값에 따른 p-value 변화',
                     labels={'x': f'{variable_name} 임계값', 'y': 'p-value'}, log_y=True)
    fig.show()

# 변수 목록 및 데이터 로드 (예시)
# data = pd.read_csv('data.csv')
variables = ['wnd_dnv_cnt', 'bus_station_cnt', 'traffic_cnt', 'crossroad_cnt', 'crosswalk_cnt']

# 변수별로 임계값 테스트 및 시각화
for var in variables:
    min_val = int(data[var].min())
    max_val = int(data[var].max())
    results = find_mann_whitney_thresholds(data, var, min_val, max_val)
    plot_threshold_results(results, var)
```

### 스피어만, 피어슨 상관계수

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import plotly.express as px
import plotly.graph_objects as go

# 유의미한 변수 
significant_vars = ['wnd_dnv_cnt', 'bus_station_cnt', 'traffic_cnt', 
                    'crossroad_cnt', '도로10_20_cnt', '도로20_30_cnt', 'child_cnt', 'silver_cnt',  'crosswalk_cnt']

def calculate_correlations(df, vars, method='pearson'):
    results = []
    for var in vars:
        if method == 'pearson':
            corr, p_value = stats.pearsonr(df[var], df['EPDO_cnt'])
        elif method == 'spearman':
            corr, p_value = stats.spearmanr(df[var], df['EPDO_cnt'])
        results.append((var, corr, p_value))
    return results

# Pearson 상관계수 계산
pearson_results = calculate_correlations(data, significant_vars, method='pearson')

# Spearman 상관계수 계산
spearman_results = calculate_correlations(data, significant_vars, method='spearman')

print("Pearson 상관계수 결과:")
for var, corr, p_value in pearson_results:
    print(f"변수 {var}와 EPDO_cnt 간의 Pearson 상관계수: {corr:.3f}, p-value: {p_value:.3e}")

print("\nSpearman 상관계수 결과:")
for var, corr, p_value in spearman_results:
    print(f"변수 {var}와 EPDO_cnt 간의 Spearman 상관계수: {corr:.3f}, p-value: {p_value:.3e}")

def plot_heatmap(corr_results, title):
    vars, corrs, p_values = zip(*corr_results)
    heatmap_df = pd.DataFrame({
        '변수': vars,
        '상관계수': corrs,
        'p-value': p_values
    }).set_index('변수')

    fig = px.imshow(heatmap_df[['상관계수']], text_auto=True, aspect="auto",
                    color_continuous_scale='RdBu_r', title=title)
    fig.update_layout(
        autosize=False,
        width=800,
        height=400,
        margin=dict(l=40, r=40, b=40, t=40),
        coloraxis_colorbar=dict(
            title="상관계수",
            titleside="right"
        )
    )
    fig.show()

# Pearson 상관계수 히트맵
plot_heatmap(pearson_results, 'Pearson 상관계수 히트맵')

# Spearman 상관계수 히트맵
plot_heatmap(spearman_results, 'Spearman 상관계수 히트맵')
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import plotly.express as px

# 기준 설정
traffic_threshold = 12
bus_station_threshold = 3
crosswalk_threshold = 6

# _threshold 열 추가
data['traffic_cnt_high'] = data['traffic_cnt'] >= traffic_threshold
data['bus_station_cnt_high'] = data['bus_station_cnt'] >= bus_station_threshold
data['crosswalk_cnt_high'] = data['crosswalk_cnt'] >= crosswalk_threshold

# 가설 테스트 
def perform_hypothesis_tests(df):
    results = []
    for col in ['traffic_cnt', 'bus_station_cnt', 'crosswalk_cnt']:
        group1 = df[df[f'{col}_high']]['EPDO_cnt']
        group2 = df[~df[f'{col}_high']]['EPDO_cnt']
        t_stat, p_value = stats.ttest_ind(group1, group2)
        results.append((col, 'EPDO_cnt', t_stat, p_value))
    return results

hypothesis_test_results = perform_hypothesis_tests(data)

for var1, var2, t_stat, p_value in hypothesis_test_results:
    print(f"변수 {var1}와 {var2} 간의 t-검정 결과: t-통계량: {t_stat:.3f}, p-value: {p_value:.3e}")

def plot_distribution(df, variable, group_variable, group1_label, group2_label):
    fig = px.histogram(df, x=variable, color=group_variable, barmode='overlay',
                       title=f'{variable} 분포 - {group1_label} vs {group2_label}')
    fig.show()

plot_distribution(data, 'EPDO_cnt', 'traffic_cnt_high', 'traffic_cnt >= 12', 'traffic_cnt < 12')
plot_distribution(data, 'EPDO_cnt', 'bus_station_cnt_high', 'bus_station_cnt >= 3', 'bus_station_cnt < 3')
plot_distribution(data, 'EPDO_cnt', 'crosswalk_cnt_high', 'crosswalk_cnt >= 6', 'crosswalk_cnt < 6')
```
